<html><b>PLOTS AND SUBPLOTS</b></html>

Q: How do I increase the size of my figure?

```python
plt.figure(figsize=(10,5))
plt.scatter(1, 1);
```

Q: What is a subplot? How do I create and navigate subplots?

```python
for i in range (1, 5):
   plt.subplot(2, 2, i)
   plt.text(0.5,0.5, str((2, 2, i)), ha='center', fontsize = 10)
```

Q: My subplot arrangement does not look very nice. How do I fix it?

We can fix it by calling plt.tight_layout(). It cleans up the padding between subplots for a cleaner look

```python
for i in range (1, 5):
   plt.subplot(2, 2, i)
   plt.text(0.5,0.5, str((2, 2, i)), ha='center', fontsize = 10)
   plt.tight_layout()
```

Q: How do I add gridlines to my plots?

```python
plt.scatter(1, 1)
plt.grid(True)
```

Q: How do I change my line colors/width/style?

You can pass in the arguments color, linewidth, and linestyle.

```python
x = [0, 2, 4, 6, 8]
y = [1, 3, 5, 7, 9]

plt.figure(figsize=(8,8))
plt.subplot(2, 2, 1)
plt.plot(x, y, color = 'red' , linewidth = 2, linestyle = ':')
plt.subplot(2, 2, 2)
plt.plot(x, y, color = 'green' , linewidth = 3, linestyle = '-.')
plt.subplot(2, 2, 3)
plt.plot(x, y, color = 'orange' , linewidth = 4, linestyle = '--')
plt.subplot(2, 2, 4)
plt.plot(x, y, color = 'blue' , linewidth = 5, linestyle = '-')
plt.tight_layout()
```

Q: How do I change my marker style or color?

You can pass in the arguments marker and color like below.

```python
m = ['^', '3', 'P', 'x', '_',
     'v', '2', 'p', '+', '|',
     'o', '1', 's', 'H', 'd',
     ',', '4', '8', 'h', 'D',
     '<', '>', '*', 'X', '']

c = ['red', 'green', 'blue', 'orange', 'purple',
     'red', 'green', 'blue', 'orange', 'purple',
     'red', 'green', 'blue', 'orange', 'purple',
     'red', 'green', 'blue', 'orange', 'purple',
     'red', 'green', 'blue', 'orange', 'purple']

for i, item in enumerate(m):
    plt.subplot(5, 5, i+1)
    plt.scatter(1, 1, marker = item, color = c[i])

plt.tight_layout()
```

Q: How do I change the transparency for my lines?

Pass in the alpha parameter to your plot. Alpha can be set from 0 to 1, where 0 is completely transparent and 1 is no transparency.

```python
plt.plot(x, y, alpha = 0.1)
```
