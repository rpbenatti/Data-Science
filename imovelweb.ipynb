```python
!pip install selenium

path = 'chromedriver.exe'

from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
```

```python
driver = webdriver.Chrome(path, chrome_options=options)
```

```python
import time

from selenium.webdriver.support.ui import Select
```

```python
url = 'https://www.imovelweb.com.br/'

driver = webdriver.Chrome(path, chrome_options=options)
driver.get(url)
time.sleep(5)

lista = Select(driver.find_element_by_xpath('//*[@id="propertyType"]'))

for item in lista.options:
    print(item.text)
    
lista.select_by_index(2)
time.sleep(5)
lista.select_by_visible_text('Sobrado')
time.sleep(5)
lista.select_by_value('10_1004')
time.sleep(5)

l = ['Casa de Condomínio','Terreno','Fazenda']

for item in l:
    lista.select_by_visible_text(item)
    time.sleep(5)
    
page_source = driver.page_source

parsed_content = BeautifulSoup(page_source, 'html.parser')
```
